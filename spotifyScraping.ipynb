```python
import pandas as pd

# Create an empty list to store the filtered song features
filtered_songs_feature_list = []

# Create a list to store the indices of elements that are removed
indices_to_remove = []

# Iterate through the song features list with index
for i, song in enumerate(song_features_list):
    if song is not None:
        filtered_songs_feature_list.append(song)
    else:
        # Append the index of the element to the list of indices to remove
        indices_to_remove.append(i)

# Convert the filtered list of JSON objects to a DataFrame
song_features_df = pd.DataFrame(filtered_songs_feature_list)

# Print the indices of elements that are removed
print("Indices of elements removed: ", indices_to_remove)
```

```python
# Remove rows with specified indices from the DataFrame
newDF = newDF.drop(indices_to_remove)

# Reset the index of the DataFrame
newDF = newDF.reset_index(drop=True)
```
