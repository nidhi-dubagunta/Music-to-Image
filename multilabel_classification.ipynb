```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x__features, y_labels, test_size=0.2, random_state=42)

print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier
br_rf = BinaryRelevance(RandomForestClassifier(n_estimators=100, verbose=2))
# train
br_rf.fit(x_train, y_train)

```

```python
predict_proba = br_rf.predict_proba(x_test)
t = 0.3
predict_new = (predict_proba >= t).astype(int)
f1_score(y_test, predict_new, average="micro")
```

```python
# using classifier chains
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import LogisticRegression
# initialize classifier chains multi-label classifier
cl_rf = ClassifierChain(RandomForestClassifier(n_estimators=100, verbose=2))
# train
cl_rf.fit(x_train, y_train)
predict_proba = cl_rf.predict_proba(x_test)
t = 0.3
predict_new = (predict_proba >= t).astype(int)
f1_score(y_test, predict_new, average="micro")



```

```python
from sklearn.metrics import roc_auc_score, precision_score, recall_score

print(roc_auc_score(y_test, predict_proba.toarray()))
```

```python
t = 0.3
predict_new = (predict_proba >= t).astype(int)
precision = precision_score(y_test, predict_new, average=None)
recall = recall_score(y_test, predict_new, average=None)
print(precision)
print(recall)
```

```python
y_cols = y_labels.columns

#create a dataframe to store the results

#columns are labels
#rows are precision and recall scores

results = pd.DataFrame(columns=['precision', 'recall'], index=y_cols)

results['precision'] = precision
results['recall'] = recall

results
```
